```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
r=3
theta=np.arange(0,2*np.pi,0.01)
f=np.linspace(-np.pi,np.pi,10)
x=r*np.sin(theta)-3
y=r*np.cos(theta)-6
a=r*np.sin(theta)+3
b=r*np.cos(theta)-6
ax = plt.gca()
ax.set_aspect('equal')
plt.xlim(-10,10)
plt.ylim(-10,20)
plt.plot(x,y)
plt.plot(a,b)
d=10-np.cosh(f)
plt.plot(f,2*d)
```
